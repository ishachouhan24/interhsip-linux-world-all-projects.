```python
import streamlit as st
import google.generativeai as genai
key = "AIzaSyCzSO7jx8cQ9IsOGsYQXGW2K_8BitQftSg"

my_info = {"Name": "Arihant Bishnoi", "Password":"987654321"}


st.header("Login Page.")

# -- session state fixes: initialize the correct keys --
if "logged_in" not in st.session_state:
    st.session_state.logged_in = False

if "name" not in st.session_state:
    st.session_state.name = ""

if "password" not in st.session_state:
    st.session_state.password = ""

if not st.session_state.logged_in:

    name = st.text_input("Enter your name : ")
    # use text_input for password so comparison with stored string works
    password = st.text_input("Enter your Password : ", type="password")
    if st.button("Login"):
        if name == my_info["Name"] and password == my_info["Password"]:
            st.session_state.name = name
            st.session_state.password = password
            st.session_state.logged_in = True
            st.success("Login Successful!")
            st.rerun()
        else:
            st.error("Invalid password or user_name")

# ---------- Chat UI (replace your current else: block with this) ----------
else:
    st.title(f"Welcome {st.session_state.name}")
    st.subheader("Hello Chatbot!")

    # configure genai (ensure key is set safely; this example uses 'key' variable)
    try:
        genai.configure(api_key=key)
        model = genai.GenerativeModel("gemini-2.5-flash")
    except Exception as e:
        st.error(f"API configuration failed: {e}")
        model = None

    # ensure messages list exists
    if "messages" not in st.session_state:
        st.session_state.messages = []  # list of tuples ("user"|"bot", text)

    st.markdown("### Conversation")
    for role, text in st.session_state.messages:
        if role == "user":
            st.markdown(f"*You:* {text}")
        else:
            st.markdown(f"*Bot:* {text}")

    # Use a form so the text input can be submitted and cleared automatically
    with st.form(key="chat_form", clear_on_submit=True):
        my_input = st.text_input("Enter what you want to search or type 'exit' to close :", key="chat_input_once")
        submit = st.form_submit_button("Send")

    if submit:
        if not my_input or not my_input.strip():
            st.warning("Please type something before sending.")
        elif my_input.strip().lower() == "exit":
            st.info("Ending chat.")
            # you can clear messages or log out
            # st.session_state.messages = []
        else:
            # append user message
            st.session_state.messages.append(("user", my_input.strip()))

            # call the API if model is configured
            if model is None:
                bot_text = "[No model configured]"
            else:
                try:
                    response = model.generate_content(
    		          my_input.strip(),
   		          generation_config={
        	              "max_output_tokens": 512
   		          }
		    )
                    bot_text = response.text
                except Exception as e:
                    bot_text = f"[API error: {e}]"

            st.session_state.messages.append(("bot", bot_text))

            # rerun so the conversation display updates (form is cleared automatically)
            st.rerun()
```
