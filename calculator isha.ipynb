```python
a=int(input("Enter your number:"))
b=int(input("enter your number:"))
print("What do you want to do?")
print("1. Add (a + b)")
print("2. Subtract (a - b)")
print("3. Multiply (a * b)")
print("4. Divide (a / b)")
operation = int(input("Enter the number of the operation (1/2/3/4): "))
if operation == 1:
    print("The result of a + b is", a + b)
elif operation == 2:
    print("The result of a - b is", a - b)
elif operation == 3:
    print("The result of a * b is", a * b)
elif operation == 4:
    if b != 0:
        print("The result of a / b is", a / b)
    else:
        print("Division by zero is not allowed.")
else:
    print("Invalid operation selected.")
    print("the number of a and b is", a-b)

```
